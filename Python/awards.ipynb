```python
import pandas as pd
from datetime import datetime

# Load the CSV file into a DataFrame
df = pd.read_csv('awards.csv')

# Convert the 'date_employed' column to datetime
df['date_employed'] = pd.to_datetime(df['date_employed'])

# Calculate the number of years spent as of the current date
current_date = datetime.now()
df['years_spent'] = (current_date - df['date_employed']).dt.days // 365

# Define the target years you want to check
target_years = [10, 15, 20, 25, 30]

# Create a result DataFrame with columns for each target year
result_df = pd.DataFrame({'Company': df['company'], 'Full Name': df['full_name']})

# Calculate the count of people who have spent exactly X years by 2024 for each X in target_years
for years in target_years:
    result_df[f'{years} Years'] = df['years_spent'] == years

# Display the result DataFrame
print(result_df)
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('awards.csv')

# Convert the 'date_employed' column to datetime using 'month/day/year' format
df['date_employed'] = pd.to_datetime(df['date_employed'], format='%m/%d/%Y')

# Calculate the number of years spent in the company by 2024
target_year = 2024
df['years_spent'] = target_year - df['date_employed'].dt.year

# Define the years you want to check
target_years = [10, 15, 20, 25, 30]

# Create a result DataFrame with columns for company, full name, and years spent
result_df = pd.DataFrame({'Company': df['company'], 'Full Name': df['full_name'], 'Years Spent': df['years_spent']})

# Loop through the target years and print the names of people who have spent exactly those years by 2024
for years in target_years:
    exact_years_df = result_df[result_df['Years Spent'] == years]
    print(f"People with exactly {years} years of employment by 2024:")
    for index, row in exact_years_df.iterrows():
        print(f"Company: {row['Company']}, Full Name: {row['Full Name']}")
    print()
```
