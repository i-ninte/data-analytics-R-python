```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the Excel file
from google.colab import files
uploaded = files.upload()

# Load the Excel file
file_name = list(uploaded.keys())[0]
data = pd.read_excel(file_name, sheet_name=None)

```

```python

# Data preprocessing
sheet1 = data['Sheet1'].dropna(how='all').reset_index(drop=True)
sheet2 = data['Sheet2']
sheet3 = data['Sheet3']

# Clean and structure Sheet1 data
columns_sheet1 = ['Extract', 'Concentration', 'ST', 'EC', 'SP', 'SA', 'CA']
sheet1.columns = columns_sheet1
```

```python

# Plotting the effects of different concentrations on organisms in Sheet2
plt.figure(figsize=(12, 8))
for organism in sheet2.columns[1:]:
    sns.lineplot(x=sheet2['Concentration of extracts'], y=sheet2[organism], label=organism)

plt.title('Effect of Extract Concentration on Test Organisms')
plt.xlabel('Concentration of Extracts (mg/ml)')
plt.ylabel('Effect on Organisms')
plt.legend(title='Test Organisms')
plt.grid(True)
plt.show()
```

As the concentration increases, the effect generally becomes more pronounced for all organisms, with a notable increase in impact at higher concentrations (50 mg/ml and above).

```python

# Visualizing the effect of different extracts on the organisms in Sheet3
plt.figure(figsize=(12, 8))
sns.heatmap(sheet3.set_index('Extract'), annot=True, cmap='viridis', fmt=".2f")
plt.title('Effect of Different Extracts on Test Organisms')
plt.ylabel('Extract')
plt.xlabel('Test Organisms')
plt.show()
```

### Analysis of the Effectiveness of Different Extracts on Test Organisms



#### 1. **Leaf Ethanol Extract**:
   - **ST (Staphylococcus aureus)**: This extract has a significant impact, showing a high level of effectiveness (16.33).
   - **EC (Escherichia coli)**: Similarly, it exhibits strong effects on EC (15.67), indicating that this extract is effective against this organism.
   - **SP (Streptococcus pneumoniae)**: The impact is also considerable here, with a value of 19.33.
   - **SA (Salmonella enterica)**: The extract is quite effective, with a recorded value of 15.67.
   - **CA (Candida albicans)**: Shows strong effectiveness (18.67).

   **Conclusion**: Across all organisms, the Leaf Ethanol Extract exhibits high levels of effectiveness, with the strongest impact on SP and CA, making it a potent antimicrobial agent in this context.

#### 2. **Stem Ethanol Extract**:
   - **ST (Staphylococcus aureus)**: Even more effective than the Leaf Ethanol Extract, with a value of 17.33.
   - **EC (Escherichia coli)**: Shows significant effects (18.67), indicating strong antimicrobial activity.
   - **SP (Streptococcus pneumoniae)**: The extracts impact is very high (19.67), almost the highest observed across all extracts and organisms.
   - **SA (Salmonella enterica)**: Again, this extract is highly effective (18.33).
   - **CA (Candida albicans)**: Strong effects are observed here as well (17.00).

   **Conclusion** : Stem Ethanol Extract is extremely effective across all test organisms, particularly against SP, ST, and EC, making it one of the most potent extracts tested.

#### 3. **Leaf Cold Water Extract**:
   - **ST (Staphylococcus aureus)**: This extract has no effect (0.00), indicating it is ineffective against this organism.
   - **EC (Escherichia coli)**: Similarly, no effect is observed (0.00).
   - **SP (Streptococcus pneumoniae)**: However, there is moderate effectiveness (14.00), suggesting some antimicrobial properties against this organism.
   - **SA (Salmonella enterica)**: It shows effectiveness (13.00), indicating it can inhibit this organism.
   - **CA (Candida albicans)**: The extract is moderately effective (17.33).

   **Conclusion**: The Leaf Cold Water Extract is generally less effective compared to ethanol extracts. It shows some activity against SP, SA, and CA, but is completely ineffective against ST and EC.

#### 4. **Stem Cold Extract**:
   - **ST (Staphylococcus aureus)**: No effect is observed (0.00).
   - **EC (Escherichia coli)**: No effect (0.00).
   - **SP (Streptococcus pneumoniae)**: No effect (0.00).
   - **SA (Salmonella enterica)**: No effect (0.00).
   - **CA (Candida albicans)**: The only organism affected, with a moderate effect (15.67).

   **Conclusion**: The Stem Cold Extract is largely ineffective against most organisms tested, except for CA, where it shows some level of activity.

#### 5. **Leaf Hot Water Extract**:
   - **ST (Staphylococcus aureus)**: No effect is observed (0.00).
   - **EC (Escherichia coli)**: It shows significant effects (15.67), indicating strong antimicrobial activity.
   - **SP (Streptococcus pneumoniae)**: No effect (0.00).
   - **SA (Salmonella enterica)**: Moderate effectiveness (13.00).
   - **CA (Candida albicans)**: Shows good effectiveness (15.33).

   **Conclusion**: The Leaf Hot Water Extract is effective against EC, SA, and CA, but has no effect on ST and SP. Its selective effectiveness suggests it could be used in specific applications where these organisms are of concern.

### **Key Takeaways:**

- **Ethanol Extracts (Leaf and Stem)**: Both show strong, broad-spectrum effectiveness against all test organisms, making them the most potent options tested.
  
- **Water Extracts**: These extracts (both cold and hot) have selective effectiveness. They are less potent overall, with the Cold Water Extracts being the least effective, showing almost no effect on most organisms except CA.

- **Organism-Specific Effects**:
  - **SP (Streptococcus pneumoniae)** and **CA (Candida albicans)** seem to be more susceptible to the extracts tested, especially to ethanol extracts.
  - **ST (Staphylococcus aureus)** and **EC (Escherichia coli)** also show significant susceptibility to the more potent extracts.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the Excel file
from google.colab import files
uploaded = files.upload()

```

```python

# Load the Excel file
file_name = list(uploaded.keys())[0]
data = pd.read_excel(file_name, sheet_name=None)

sheet1 = data['Sheet1'].dropna(how='all').reset_index(drop=True)
sheet2 = data['Sheet2']
sheet3 = data['Sheet3']

# Clean and structure Sheet1 data
columns_sheet1 = ['Extract', 'Concentration', 'ST', 'EC', 'SP', 'SA', 'CA']
sheet1.columns = columns_sheet1

```

```python
# Calculate the correlation matrix for Sheet2
corr_matrix_sheet2 = sheet2.corr()
```

```python

# Plot the correlation matrix for Sheet2
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_sheet2, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Sheet2')
plt.show()
```



The high correlations between SP, ST, and EC suggest that the extracts generally work consistently across these organisms. If an extract is effective on one, it is likely to be effective on the others.
Effect of Concentration:

The correlation between the concentration of extracts and their effectiveness on SP, EC, and ST suggests that increasing the concentration generally enhances their effectiveness on these organisms.
Selective Effectiveness:

The relatively lower correlation between the concentration of extracts and SA suggests that increasing the concentration may not be as effective in improving outcomes against SA compared to other organisms. Similarly, the lower correlation between CA and other organisms suggests some degree of selectivity in effectiveness.
Potential for Targeted Application:

Given the different levels of correlation, extracts could potentially be optimized for specific organisms by adjusting concentrations. For instance, a higher concentration might be particularly beneficial for treating infections caused by SP or EC.
